```python
# Instalar transformers e sentencepiece (necessário para alguns modelos)
!pip install -q transformers sentencepiece

from transformers import pipeline

# Criar o pipeline de tradução (inglês -> francês)
translator = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr")

# Texto para traduzir
text = "Hello world"

# Executar tradução
result = translator(text, max_length=40)

# Exibir resultado
print("Tradução:", result[0]['translation_text'])
```

```python
def agent_run(command: str):
    command = command.lower()
    if "calculate" in command:
        # Extrair a expressão após "calculate"
        # E antes de "in python", se existir
        start = command.find("calculate") + len("calculate")
        end = command.find("in python")
        if end == -1:
            # Se não tiver "in python", pega até o final
            expression = command[start:].strip()
        else:
            expression = command[start:end].strip()

        try:
            result = eval(expression, {}, {})
            return str(result)
        except Exception as e:
            return f"Erro na avaliação: {e}"
    else:
        return "Comando não entendido."

# Teste
comando = "Calculate 2+2 in Python"
resposta = agent_run(comando)
print("Resposta do agente:", resposta)
```
